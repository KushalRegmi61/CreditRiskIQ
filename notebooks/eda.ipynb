```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/cs-training.csv')


# Drop unnamed index column if it exists
if 'Unnamed: 0' in df.columns:
    df.drop(columns='Unnamed: 0', inplace=True)

# Sample 20,000 rows with a fixed random seed for reproducibility
df_sampled = df.sample(n=40000, random_state=42).reset_index(drop=True)

# Check basic info
print(df_sampled.shape)
df_sampled.head()
```

```python
df_sampled.to_csv('../data/cs-sampled.csv', index=False)
```
